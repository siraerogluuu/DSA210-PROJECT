# TikTok Video Performance Analysis
This project analyzes the relationship between engagement metrics such as likes, comments, and views in TikTok videos. The goal is to understand how user interaction affects video performance.

## Data Loading

```python
import pandas as pd

df = pd.read_csv("tiktok_performance.csv")
df.head()
```

```python
plt.scatter(df['Comments'], df['Views'])
plt.show()
```

```python
df.groupby('Category')['Views'].mean()
```

```python
df.groupby('Category')['Views'].mean().plot(kind='bar')
plt.show()
```

```python
df['engagement'] = df['Likes'] + df['Comments'] + df['Shares']
```

```python
plt.scatter(df['engagement'], df['Views'])
plt.show()
```

```python
df[['Likes','Comments','Views']].corr()
```

## Hypothesis

Videos with higher numbers of likes and comments are expected to receive higher view counts.

## Conclusion

The analysis shows that higher engagement (likes and comments) is associated with higher view counts. It indicates that user interaction plays a key role in the popularity of TikTok videos. Moreover, this supports the hypothesis that user engagement significantly influences video visibility.
